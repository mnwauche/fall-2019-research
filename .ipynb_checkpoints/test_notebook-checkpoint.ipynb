```python
def gnet_slow_forward(gnet, x):
        interm_results = []
        
        ## Normalize the input ?
        if gnet.normalizationLayer is not None:
            x = gnet.normalizationLayer(x)
            interm_results.append((x, 'normalization'))
        x = x.view(-1, num_flat_features(x))
        interm_results.append((x, 'normalization'))
        
        # format layer
        x = gnet.leakyRelu(self.formatLayer(x))
        interm_results.append((x, 'format'))
        x = x.view(x.size()[0], -1, 4, 4)
        x = gnet.normalizationLayer(x)
        interm_results.append((x, 'format'))
        
        # Scale 0 (no upsampling)
        for convLayer in gnet.groupScale0:
            x = gnet.leakyRelu(convLayer(x))
            if gnet.normalizationLayer is not None:
                x = gnet.normalizationLayer(x)
            interm_results.append((x, 'scale0'))
            
        # Dirty, find a better way
        if gnet.alpha > 0 and len(gnet.scaleLayers) == 1:
            y = gnet.toRGBLayers[-2](x)
            y = Upscale2d(y)
            interm_results.append((x, 'scale1'))

        # Upper scales
        for scale, layerGroup in enumerate(gnet.scaleLayers, 0):

            x = Upscale2d(x)
            interm_results.append((x, 'upper_scales0'))
            
            for convLayer in layerGroup:
                x = gnet.leakyRelu(convLayer(x))
                if gnet.normalizationLayer is not None:
                    x = gnet.normalizationLayer(x)
                    
                interm_results.append((x, 'upper_scales1(layerGroup)'))

            if gnet.alpha > 0 and scale == (len(gnet.scaleLayers) - 2):
                y = gnet.toRGBLayers[-2](x)
                y = Upscale2d(y)

                interm_results.append((y, 'upper_scales2(y)'))


        # To RGB (no alpha parameter for now)
        x = gnet.toRGBLayers[-1](x)
        interm_results.append((x, 'toRGB'))

        # Blending with the lower resolution output when alpha > 0
        if self.alpha > 0:
            x = self.alpha * y + (1.0-self.alpha) * x
            

        if self.generationActivation is not None:
            x = self.generationActivation(x)
        interm_results.append((x, 'after_blending'))

        return x, interm_results
```
